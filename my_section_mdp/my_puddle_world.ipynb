```python
class Goal: #ゴールの設定，Landmarkクラスを改変
    def __init__(self, x, y, radius=0.3, value=0.0):
        self.pos = np.array([x, y]).T
        self.radius = radius
        self.value = value #ゴールの終端状態としての価値
        
    def inside(self, pose): #ロボットがゴールに存在するかどうかを判断
        return self.radius > math.sqrt( (self.pos[0]-pose[0])**2 + (self.pos[1]-pose[1])**2 )
        
    def draw(self, ax, elems):
        x, y = self.pos
        c = ax.scatter(x + 0.16, y + 0.5, s=50, marker=">", label="landmarks", color="red") #旗
        elems.append(c)
        elems += ax.plot([x, x], [y, y + 0.6], color="black") #旗竿
```

```python
class Puddle: #puddle
    def __init__(self, lowerleft, upperright, depth):
        self.lowerleft = lowerleft
        self.upperright = upperright
        self.depth = depth #水たまりの深さ，深いほどロボットに対して害がある
        
    def draw(self, ax, elems):
        w = self.upperright[0] - self.lowerleft[0]
        h = self.upperright[1] - self.lowerleft[1]
        r = patches.Rectangle(self.lowerleft, w, h, color="blue", alpha=self.depth)
        elems.append(ax.add_patch(r))
    
    def inside(self, pose): #水たまりの中か外かを判断する
        return all([ self.lowerleft[i] < pose[i] < self.upperright[i]  for i in [0, 1] ])
```

```python
class PuddleWorld(World):
    def __init__(self, time_span, time_interval, debug=False):
        super().__init__(time_span, time_interval, debug)
        self.puddles = [] #水たまりのオブジェクトを登録するリスト
        self.robots = [] #robotを登録するリスト
        self.goals = [] #ゴールを登録するリスト
        
    def append(self,obj):
        self.objects.append(obj)
        if isinstance(obj, Puddle): self.puddles.append(obj) #objとPuddleが等しければ追加
        if isinstance(obj, Robot): self.robots.append(obj)
        if isinstance(obj, Goal): self.goals.append(obj)
        
    def puddle_depth(self, pose): #水たまりの深さを計算する関数
        return sum([p.depth * p.inside(pose) for p in self.puddles])
    
    def one_step(self, i, elems, ax):
        super().one_step(i, elems, ax)
        for r in self.robots:
            r.agent.puddle_depth = self.puddle_depth(r.pose)
            for g in self.goals:
                if g.inside(r.pose): #ロボットがゴールに存在していたら
                    r.agent.in_goal = True #in_goalをTrueにする
                    r.agent.final_value = g.value #finalvalueを設定
```

```python
class PuddleIgnoreAgent(EstimationAgent):#\
    def __init__(self, time_interval, estimator, goal, puddle_coef=100):
        super().__init__(time_interval, 0.0, 0.0, estimator)
        
        self.puddle_coef = puddle_coef #cに当たる変数の設定
        self.puddle_depth = 0.0 #水たまりの深さ
        self.total_reward = 0.0 #これまでの報酬の総和
        self.in_goal = False #現在ゴールに存在するかどうか
        self.final_value = 0.0 #終端状態の価値
        self.goal = goal #ゴールの把握
        
    def reward_per_sec(self): #その瞬間の評価関数の値を返す
        return -1.0 - self.puddle_depth*self.puddle_coef
    
    @classmethod
    def policy(cls, pose, goal): #ゴールの方向を向いて直進する単純な方策を実装
        x, y, theta = pose #現在の姿勢を保存
        dx, dy = goal.pos[0] - x, goal.pos[1] - y #ゴールとの差を計算
        direction = int((math.atan2(dy, dx) - theta)*180/math.pi) #ゴールの方角を計算
        direction = (direction + 360*1000 + 180)%360 - 180 #方角を-180〜180[deg]に正規化
        
        if direction > 10: nu, omega = 0.0, 2.0 #方角が10[deg]以上ずれていたら回転
        elif direction < -10: nu, omega = 0.0, -2.0
        else: nu, omega = 1.0, 0.0 #方角のずれが10[deg]以下なら直進
            
        return nu, omega

    def decision(self, observation=None): #報酬を積算する関数
        if self.in_goal: #ゴールしたら停止する
            return 0.0, 0.0
        
        self.estimator.motion_update(self.prev_nu, self.prev_omega, self.time_interval) #カルマンフィルタの処理(移動)
        self.estimator.observation_update(observation) #カルマンフィルタの処理(観測)
        
        self.total_reward += self.time_interval*self.reward_per_sec() #報酬の計算

        nu, omega = self.policy(self.estimator.pose, self.goal) #poclicyメソッドを利用して行動決定する
        self.prev_nu, self.prev_omega = nu, omega
        return nu, omega
        
    def draw(self, ax, elems): 
        super().draw(ax, elems)
        x, y, _ = self.estimator.pose
        elems.append(ax.text(x+1.0, y-0.5, "reward/sec:" + str(self.reward_per_sec()), fontsize=8))
        elems.append(ax.text(x+1.0, y-1.0, "eval: {:.1f}".format(self.total_reward+self.final_value), fontsize=8)) #終端状態の価値と報酬の和を表示
```

```python
def trial():  ###puddle_world4_trial
    time_interval = 0.1
    world = PuddleWorld(30, time_interval, debug=False) 

    ## 地図を生成して3つランドマークを追加 ##
    m = Map()
    for ln in [(-4,2), (2,-3), (4,4), (-4,-4)]: m.append_landmark(Landmark(*ln))
    world.append(m)   

    ##ゴールの追加##
    goal = Goal(-3,-3)  #goalを変数に
    world.append(goal)
    
    ##水たまりの追加##
    world.append(Puddle((-2, 0), (0, 2), 0.1)) 
    world.append(Puddle((-0.5, -2), (2.5, 1), 0.1)) 

   ##ロボットを作る##
    initial_pose = np.array([2, 2,0]).T
    kf = KalmanFilter(m, initial_pose) 
    a = PuddleIgnoreAgent(time_interval, kf, goal) #goalを渡す
    r = Robot(initial_pose, sensor=Camera(m, distance_bias_rate_stddev=0, direction_bias_stddev=0), 
              agent=a, color="red", bias_rate_stds=(0,0))
    world.append(r)
    
    world.draw()
    
trial()
```
