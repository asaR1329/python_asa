```python
world = World(40.0, 0.1)      ###motion_test_forward_bias

initial_pose = np.array([0, 0, 0]).T
robots = []

for i in range(100):
    r = Robot(initial_pose, sensor=None, agent=Agent(0.1, 0.0))  #ã“ã“ã§ç”Ÿæˆã•ã‚Œã‚‹ãƒ­ãƒœãƒƒãƒˆã¯ç•°ãªã‚‹ãƒã‚¤ã‚¢ã‚¹ã‚’æŒã¤
    world.append(r)
    robots.append(r)

world.draw()
```
